### Iteration and Function Applications

###### function application

```python
df
```

```python
df['Full_address'] = df[['Address','City','Country']].apply(lambda x: '{} in {}, {}'.format(x['Address'].split()[0],x['City'],x['Country']),axis=1)
df
```

iteration
1. iteritems
2. iterrows
3. itertuples

```python
for i in df:
    print(df[i])
```

```python
for i in range(1,len(df)+1):
    print(df.loc[i])
```

```python
k = []
for i in range(1,len(df)+1):
    k.append('{} in {}, {}'.format(df['Address'][i].split()[0],df['City'][i],df['Country'][i]))
print(k)
```

```python
for i,v in df.iteritems():
    print(i)
    print(v)
    break
```

```python
for i,v in df.iterrows():
    print(i)
    print(v)
    break
```

```python
for i in df.itertuples():
    print(i)
    print(i[1])
```
