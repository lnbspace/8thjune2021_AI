<b>Graphs</b>
1. plot - line plotting
2. bar - bar graph of data
    - barh - horizontal version of bar
3. hist - histogram of data (frequency of data)
4. scatter - point plot of realtionship between two features
5. boxplot - boxplot (median,0.25/0.75 quantile, Outliers)
6. box - same as boxplot
7. stem - point plot with bars
8. violinplot - density of data in the shape of violin

```python
a = np.random.randint(1,100,(30))
b = np.random.randint(1,100,(30)) * np.random.rand(1)
```

```python
plt.plot(a)
```

```python
plt.plot(b)
```

```python
plt.figure(figsize=(10,6))
plt.plot(a, color='r', ls='-.',marker='*',markersize=10,markerfacecolor='blue',markeredgecolor='g')
plt.grid()
plt.title('My Data')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.xticks(range(0,31,2),list('abcdefghijklmnop'))
plt.yticks(range(0,101,5))
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(a, color='r', ls='-.',marker='*',markersize=10,markerfacecolor='blue',markeredgecolor='g')
plt.plot(b, color='m', ls='--',marker='o',markersize=10,markerfacecolor='red',markeredgecolor='blue')
plt.grid()
plt.title('My Data')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.xticks(range(0,31,2),list('abcdefghijklmnop'))
plt.yticks(range(0,101,5))
plt.legend(['Data A','Data B'])
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.plot(a, color='r', ls='-.',marker='*',markersize=10,markerfacecolor='blue',markeredgecolor='g')
plt.grid()
plt.title('My Data A')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.xticks(range(0,31,2),list('abcdefghijklmnop'))
plt.yticks(range(0,101,5))
plt.legend(['Data A'])

plt.subplot(1,2,2)
plt.plot(b, color='m', ls='--',marker='o',markersize=10,markerfacecolor='red',markeredgecolor='blue')
plt.grid()
plt.title('My Data B')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.xticks(range(0,31,2),list('abcdefghijklmnop'))
#plt.yticks(range(0,101,5))
plt.legend(['Data B'])

plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.boxplot(a, patch_artist=True,showmeans=True)
plt.grid()
plt.title('My Data A')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.yticks(range(0,101,5))
plt.legend(['Data A'])

plt.subplot(1,2,2)
plt.plot(b, color='m', ls='--',marker='o',markersize=10,markerfacecolor='red',markeredgecolor='blue')
plt.grid(which='both')
plt.title('My Data B')
plt.xlabel('x axis -->')
plt.ylabel('Values -->')
plt.xticks(range(0,31,2),list('abcdefghijklmnop'))
#plt.yticks(range(0,101,5))
plt.legend(['Data B'])

plt.show()
```

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_axes([0.01,0.01, 0.95,0.95])
ax2 = fig.add_axes([0.1,0.1,0.4,0.4])
ax3 = fig.add_axes([0.6,0.5,0.2,0.2])
ax1.plot(a)
ax2.plot(y)
plt.show()
```
