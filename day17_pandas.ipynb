### Data Grouping

```python
df = pd.read_csv('Data.csv')
df
```

```python
df['Age'][df['Country']=='France'].mean()
```

```python
for c in df['Country'].unique():
    print(c,df['Age'][df['Country']==c].mean())
```

```python
d1 = pd.DataFrame({'name':['Rahul','Vishesh','Kanika'], 'age':[23,22,20],'uid':[101,102,103]})
d1
```

```python
d2 = pd.DataFrame({'subject':['python','java','cpp','php'], 'uid':[103,101,104,105]})
d2
```

```python
pd.merge(d1,d2, on='uid') #inner
```

```python
pd.merge(d1,d2, on='uid', how='outer')
```

```python
pd.merge(d1,d2, on='uid', how='left')
```

```python
pd.merge(d1,d2, on='uid', how='right')
```
