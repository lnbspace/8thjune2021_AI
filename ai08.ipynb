```python
(xtrain,ytrain),(xtest,ytest) = cifar10.load_data()
```

```python
xtrain.shape
```

```python
from keras.utils.np_utils import to_categorical
```

```python
ytrain_c = to_categorical(ytrain)
ytest_c = to_categorical(ytest)
```

```python
x = keras.layers.Input((32,32,3))
num_classes = len(set(ytrain.T[0].tolist()))
```

```python
model = keras_resnet.models.ResNet152(x, classes = num_classes)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain_c,verbose=1)
```
